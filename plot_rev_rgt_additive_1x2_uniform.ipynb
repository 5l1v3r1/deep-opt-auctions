```python
from __future__ import absolute_import
from __future__ import division
from __future__ import print_function

import os
import sys
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = []
fq = 40000
for i in np.arange(int(400000/fq) + 1):
    ri = i * fq
    f = open("experiments/additive_1x2_uniform/test_iter_" + str(ri) + "_m_1000_gd_2000.txt")
    data.append(f.readlines()[-1].split(",")[2:5])
```

```python
data = np.array(data)
rev = [float(d.split(" ")[-1][:-1]) for d in data[:,0]]
rgt = [float(d.split(" ")[-1][:-1]) for d in data[:,1]]
epochs = np.arange(len(rev))*(fq/5000)
```

```python
plt.rcParams.update({'font.size': 10, 'axes.labelsize': 'x-large'})
fig, ax = plt.subplots(ncols = 1, nrows = 1, figsize = (4,3))

line1, = plt.plot(epochs, rev, linewidth = 2.5, color='blue', marker ='.')
line2, = plt.plot(epochs, [0.55]*len(epochs), color='green', linewidth = 2.5)
plt.ylim([0.4, 0.6])
plt.yticks([0.4, 0.45, 0.5, 0.55, 0.6])
plt.xlabel("Epochs")
plt.ylabel("Test Revenue")

plt.legend((line1, line2), ('RegretNet', 'Optimal\nMechanism'), prop={'size': 15})
plt.savefig('rev_vs_epoch.pdf', bbox_inches = 'tight', pad_inches = 0.05)
```

```python
plt.rcParams.update({'font.size': 10, 'axes.labelsize': 'x-large'})
fig, ax = plt.subplots(ncols = 1, nrows = 1, figsize = (4,3))

line1, = plt.plot(epochs, rgt, linewidth = 2.5, color='blue', marker ='.')
plt.ylim([0,1e-2])
plt.xlabel("Epochs")
plt.ylabel("Test Regret")
plt.savefig('rgt_vs_epoch.pdf', bbox_inches = 'tight', pad_inches = 0.05)
```
